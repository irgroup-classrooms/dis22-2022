```python
cluster_frequency = {}
relevant_clusters = [18,19,243,370,21,88,398,517,613,690]
for row in merged_inner.itertuples(index=True, name='Pandas'):
    #allgemein
    if row.cluster in relevant_clusters:
        if row.occupation in cluster_frequency:
            if row.cluster in cluster_frequency[row.occupation]['neutral']:
                cluster_frequency[row.occupation]['neutral'][row.cluster] += 1
            else:
                cluster_frequency[row.occupation]['neutral'][row.cluster] = 1
            if row.gender in cluster_frequency[row.occupation]:
                if row.cluster in cluster_frequency[row.occupation][row.gender]:
                    cluster_frequency[row.occupation][row.gender][row.cluster] += 1
                else:
                    cluster_frequency[row.occupation][row.gender][row.cluster] = 1
            else:
                cluster_frequency[row.occupation][row.gender] = {}
                cluster_frequency[row.occupation][row.gender][row.cluster] = 1
        else:
            cluster_frequency[row.occupation] = {}
            cluster_frequency[row.occupation]['neutral'] = {}
            cluster_frequency[row.occupation]['neutral'][row.cluster] = 1    
print(cluster_frequency)
#neutral = beide Geschlechter
```

```python
#Quelle: https://nbviewer.org/github/bmabey/pyLDAvis/blob/master/notebooks/pyLDAvis_overview.ipynb#topic=0&lambda=1&term=
#Quelle: https://github.com/bmabey/pyLDAvis/issues/4
#!pip install pyLDAvis
from imp import reload
#from importlib import reload
import pyLDAvis
import gensim
from gensim import corpora, models, similarities
import pyLDAvis.gensim as gensimvis
import pyLDAvis
import sys
import re
from collections import defaultdict

documents = merged_inner[(merged_inner["occupation"] == 'Entertainment') & (merged_inner["gender"] == 'männlich')]['tokens'][:10000]
print(documents)
stoplist = set('for a of the and to in'.split())
texts = [[word for word in document.split() if word not in stoplist]
         for document in documents]
cleaned_texts = []
for text in texts:
    cleaned_tokens = []
    for token in text:
        token = re.sub(r'[\[\]\']', '', token)
        cleaned_tokens.append(token)
    cleaned_texts.append(cleaned_tokens)
texts = cleaned_texts
frequency = defaultdict(int)
for text in texts:
    for token in text:
        frequency[token] += 1
texts = [[token for token in text if frequency[token] > 1]
         for text in texts]
texts = [ele for ele in texts if ele != []]
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
lda = models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=30, passes=50)
visdata = gensimvis.prepare(lda, corpus, dictionary)
pyLDAvis.display(visdata)
```
